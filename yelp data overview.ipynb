```python
import pandas as pd

path = "yelp_academic_dataset_business.json"

df_business = pd.read_json(
    path,
    lines=True
)
```

```python
from collections import Counter
attr_counter = Counter()

for attrs in df_business['attributes'].dropna():
    for k in attrs.keys():
        attr_counter[k] += 1
```

```python
attr_counter.most_common()
```

```python
def flatten_attributes(attrs):
    flat = {}
    for k, v in attrs.items():
        if isinstance(v, dict):
            for sub_k, sub_v in v.items():
                flat[f"{k}_{sub_k}"] = sub_v
        else:
            flat[k] = v
    return flat
df_flat = df_business['attributes'].dropna().apply(flatten_attributes)
flat_counter = Counter()

for d in df_flat:
    for k in d:
        flat_counter[k] += 1

```

```python
flat_counter.most_common()
```

```python
df_user['yelping_since'] = pd.to_datetime(df_user['yelping_since'])
df_user.isna().sum()
```

```python
df_user['review_count'].describe()
# mean >> median 
# skewed distribution
```

```python
df_user['tenure_years'] = (
    pd.Timestamp.now() - df_user['yelping_since']
).dt.days / 365
df_user['tenure_years'].describe()
```

```python
df_user['is_elite'] = df_user['elite'].apply(lambda x: 0 if x=='' else 1)
df_user['is_elite'].value_counts(normalize=True)
# elite的用户评论权重更高？
```

```python
compliment_cols = [c for c in df_user.columns if c.startswith('compliment_')]
df_user['total_compliments'] = df_user[compliment_cols].sum(axis=1)
df_user['total_compliments'].describe()
# value = 0 做冷启动？ 
```
